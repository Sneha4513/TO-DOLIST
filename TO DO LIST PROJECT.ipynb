```python
import os

def show_task(task):
    if not task:
        print("no task found")
    else:
        for i, t in enumerate(task, 1):
            print(f"{i}. {t}")

def add_task(task, new_task):
    task.append(new_task)
    print("Task added successfully")

def update_task(task, index, updated_task):
    if 1 <= index <= len(task):
        task[index - 1] = updated_task
        print("Task updated successfully")
    else:
        print("Invalid task index")

def delete_task(task, index):
    if 1 <= index <= len(task):
        deleted_task = task.pop(index - 1)
        print(f"Task '{deleted_task}' deleted successfully")
    else:
        print("Invalid task index")

def save_task_to_file(file_path, task):
    with open(file_path, 'w') as file:
        for t in task:
            file.write(f"{t}\n")

def load_task_from_file(file_path):
    task = []
    if os.path.exists(file_path):
        with open(file_path, 'r') as file:
            task = file.read().splitlines()
    return task

def main():
    file_path = "file_path.txt"
    task = load_task_from_file(file_path)
    while True:  # Changed from 'true' to 'True'
        print("\n==== To Do List ====")
        print("1. Show Tasks")
        print("2. Add Task")
        print("3. Update Task")
        print("4. Delete Task")
        print("5. Save and Exit")
        choice = input("Enter your choice (1-5): ")
        if choice == '1':
            show_task(task)
        elif choice == '2':
            new_task = input("Enter task to add: ")
            add_task(task, new_task)
        elif choice == '3':
            index = int(input("Enter the task index: "))
            updated_task = input("Enter the updated task: ")
            update_task(task, index, updated_task)
        elif choice == '4':
            index = int(input("Enter the task index: "))
            delete_task(task, index)
        elif choice == '5':
            save_task_to_file(file_path, task)
            print("Tasks saved.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
